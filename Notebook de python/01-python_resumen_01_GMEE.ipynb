# <span style="color:#0485CF"> Condicionales: if </span>

- Las sentencias condicionales nos dan la oportunidad de cambiar el comportamiento de un programa de acuerdo a ciertos criterios.

- La sentencia condicional más sencilla es el **if**
- La expresión booleana que se coloca después del **if** se le conoce como la **condición**; 
    - si esta condición es verdadera, el código que le sigue al **if** es ejecutado 
    - sino, no pasa nada.
- No hay límites en cuanto al número de sentencias que pueden ir dentro de un **if**, pero si debe de existir al menos una.

````python
x = 1
if x > 0:
    print(x,"Mayor que 0")

````

## Condicionales encadenados
- En ocasiones solo dos posibilidades no son suficientes, en estos casos necesitamos el condicional **elif**.

- **elif** es una abreviatura de **elseif**
- No hay límite en el número de **elif** que se pueden utilizar, pero la última opción debe terminar con un **else**
- De las múltiples opciones solo una se ejecutará.
- Cada condición se analizará en orden 
    - si la primera es falsa se analizará la siguiente 
    - así hasta encontrar la condición que se cumple o no.
- Aunque más de una condición sea verdadera se ejecutará solo la primera.

````python
x , y= 99, 19
if x < y:
    print(x, "es menor que", y)
elif x > y:
    print(x, "es mayor que", y)
else:
    print(x, " y ", y, " son iguales")
````

## Condiciones anidadas
- Una condición puede estar anidada en otra.

````python
x , y = 2 , 3
if x == y:
    print(x, "Y", y, "son iguales")
else:
    if x < y:
        print(x, "es menor que", y)
    else:
        print(x, "es mayor que", y)
````

- Nombre_Funcion(lista_de_argumentos).

- Si bien, Python tiene muchas funciones pre-definidas, también es posible crear nuestras propias funciones.

## Sintaxis: Funciones
La sintaxis para crear una función es:

<br/>

- Estructura



## Funciones: Argumentos de longitud variable

- Si declaramos un argumento con * esto hará que el argumento que se pase sea empaquetado en una tupla de manera automática

- Los valores de cada argumento son posicionales, el primero corresponde a numeros[0], y así sucesivamente

- No confundir **\*** con los punteros en otros lenguajes de programación, no tiene nada que ver.

 ```python:
    def suma(*numeros):
        print(type(numeros))
        # <class 'tuple'>  similar a una lista
        print(numeros)

    suma(1,3,5,4)
 ```


````python
def suma(*numeros):
    print(type(numeros))
    # <class 'tuple'>  similar a una lista
    print(numeros)
    print(numeros[3])


suma(1, 3, 5, "hola") 
````

 ```python:
def suma(**kwargs):

suma(a=10, b=20, c=3)
 ```

## Bucle For
- La sentencia **for** forma un bucle definido que recorre los elementos de una colección de datos:  listas, diccionarios, tuplas, strings, etc. 

- Tienen un número **definido** de iteraciones.

- El bucle hará tantas iteraciones cuántos elementos hay en la secuencia.


- Estructura

 ```python:
    for <variable> in <secuencia>:
        <bloque de código>
   
 ```

#### Ejemplo

````python

idiomas = ['Árabe', 'Inglés', 'Francés', 'Español']
indice = 0
for idioma in idiomas:
    print ('Este idioma está en la lista: ' + idiomas[indice])
    indice += 1
````

### Implemente un programa que solicite al usuario un intervalo (límite inferior y límite superior, validar los límites) para evaluar  las siguientes funciones $f(x)$  para la variable $x$; así como la derivada de la función $f'(x)$

#### Instrucciones:

- ##### 1. Implementar como una función de Python el cálculo de la función $f(x)$ 

- ##### 2. Implementar como una función de Python la derivada analítica de $f(x)$, es decir, implementar la función de Python para $f'(x)$

- ##### 3. Implementar un programa principal que calcule la tabla de resultados de la evaluación de la funcion  $f(x)$   y su derivada $f'(x)$ en el intervalo proporcionado por el usuario.

### I. Funcion Polinómica

$$
\Large f(x) = 3x^4 - 5x^2 + 2x - 7 
$$

### II. Funcion exponencial

$$
\Large f(x) = e^{2x}
$$

### III. Funcion exponencial

$$
\Large f(x) = e^{-3x^2} 
$$



````python
import math
def calcular_f1(x):
    res = (3*x**4) - (5*x**2) + (2*x) -7
    return res
def calcular_f2(x):
    res = math.e**(2*x)
    return res
def calcular_f3(x):
    res = math.e**(-3*x**2)
    return res
def derivada_f1(x):
    res = (12*x**3) - (10*x) + 2
    return res
def derivada_f2(x):
    res = 2*math.e**(2*x)
    return res
def derivada_f3(x):
    res = (-6*x)*math.e**(-3*x**2)
    return res
````

````python
print("Primera funcion")
Iini=int(input("¿Cuál es el numero de intervalo incial?\n"))
Ifin=int(input("¿Cuál es el numero de intervalo final?\n"))

print("f(x)\t|\tf´(x)")
for x in range(Iini,Ifin+1):
    print(calcular_f1(x),"\t|\t",derivada_f1(x))

print("Segunda funcion")
Iini=int(input("¿Cuál es el numero de intervalo incial?\n"))
Ifin=int(input("¿Cuál es el numero de intervalo final?\n"))

print("f(x)\t|\tf´(x)")
for x in range(Iini,Ifin+1):
    print(calcular_f2(x),"\t|\t",derivada_f2(x))

print("Tercera funcion")
Iini=int(input("¿Cuál es el numero de intervalo incial?\n"))
Ifin=int(input("¿Cuál es el numero de intervalo final?\n"))

print("f(x)\t|\tf´(x)")
for x in range(Iini,Ifin+1):
    print(calcular_f3(x),"\t|\t",derivada_f3(x))
````

- ### 3. Implementa un programa que normalice los datos (normalización Min-Max) de una lista de entrada dentro de un intervalo proporcionado. La normalización de datos está dada por lo siguiente.


    - #### Normalización Min-Max

        #### La normalización lineal transforma los datos al intervalo `[a, b]` mediante:

        $$
        \boxed{
        \Large x_{\text{normalizado}} = a + \frac{(x - x_{\min}) \times (b - a)}{x_{\max} - x_{\min}}
        }
        $$

        #### Donde:
        - #### $ x $: Valor original
        - #### $ x_{\min} $: Valor mínimo en el conjunto original
        - #### $ x_{\max} $: Valor máximo en el conjunto original
        - #### $ a $: Límite inferior del rango objetivo (típicamente 0)
        - #### $ b $: Límite superior del rango objetivo (típicamente 1)

        - ### Casos Especiales

            - #### 1. **Cuando $ x_{\max} = x_{\min} $** (todos los valores iguales):
                $$
                \Large x_{\text{normalizado}} = \frac{a + b}{2}
                $$

            - #### 2. **Normalización estándar [0, 1]** (caso particular común):
                $$
                \Large x_{\text{normalizado}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
                $$

        - #### Transformación Inversa

            - #### Para recuperar el valor original:
            $$
            \Large x = x_{\min} + \frac{(x_{\text{normalizado}} - a) \times (x_{\max} - x_{\min})}{b - a}
            $$


       ####  Ejemplo de uso

        ```python
        # Normalizar al rango [-5, 5]
        datos_normalizados, min_orig, max_orig = normalizar_lista(datos_originales, -5, 5)
        ```


````python
def minimo(lista):
    if lista == []:
        return None
    menor = lista[0]
    i=1
    while i < len(lista):
        if menor > lista[i]:
            menor = lista[i]
        i=i+1
    return menor

def maximo(lista):
    if lista == []:
        return None
    maximo = lista[0]
    i=1
    while i < len(lista):
        if maximo < lista[i]:
            maximo = lista[i]
        i=i+1
    return maximo
````

````python
def normalizar_lista(lista, a=0, b=1):
    # Si la lista está vacía, regresamos valores por defecto
    if not lista:
        return [], None, None
    
    # Encontramos el mínimo y máximo de la lista
    x_min = minimo(lista)
    x_max = maximo(lista)
    
    normalizados = []

    # Caso especial: todos los valores son iguales
    if x_min == x_max:
        for _ in lista:
            normalizados.append((a + b) / 2)  # mismo valor para todos
        return normalizados, x_min, x_max

    # Normalización Min-Max
    for x in lista:
        x_norm = a + ((x - x_min) * (b - a)) / (x_max - x_min)
        normalizados.append(x_norm)

    return normalizados, x_min, x_max
````

````python
def desnormalizar_lista(lista_normalizada, x_min, x_max, a=0, b=1):
    originales = []

    for x_norm in lista_normalizada:
        x = x_min + ((x_norm - a) * (x_max - x_min)) / (b - a)
        originales.append(x)

    return originales
````

````python
#PRIMER CASO DE PRUEBA: cuando la lista esta vacia
datos_originales = []
print("Datos originiales: ",datos_originales)
#Normalización al rango [-5, 5]
rang_inf = -5
rang_sup = 5
datos_normalizados, min_orig, max_orig = normalizar_lista(datos_originales, rang_inf,rang_sup)
print("Rango [{},{}]".format(rang_inf,rang_sup))
print("Valor mínimo:", min_orig)
print("Valor máximo:", max_orig)
print("Datos normalizados:", datos_normalizados)
````

````python
#SEGUNDO CASO DE PRUEBA: cuando todos los  datos son iguales
datos_originales = [10,10,10,10]
print("Datos originiales: ",datos_originales)
#Normalización al rango [-5, 5]
rang_inf = -5
rang_sup = 5
datos_normalizados, min_orig, max_orig = normalizar_lista(datos_originales, rang_inf,rang_sup)
print("Rango [{},{}]".format(rang_inf,rang_sup))
print("Valor mínimo:", min_orig)
print("Valor máximo:", max_orig)
print("Datos normalizados:", datos_normalizados)
````

````python
# TERCER CASO DE PRUEBA: normalizacion Min-Max
datos_originales = [150, 160, 170, 180, 190]
print("Datos originiales: ",datos_originales)

#Normalización al rango [-5, 5], el caso tipico es cambiar estos valores oor [0,1]
rang_inf = -5
rang_sup = 5
#Si cambiamos EL RANGO de  uso tipico entra al mismo caso que el de Normalizacion Min-Max
datos_normalizados, min_orig, max_orig = normalizar_lista(datos_originales, rang_inf,rang_sup)
print("Rango [{},{}]".format(rang_inf,rang_sup))
print("Valor mínimo:", min_orig)
print("Valor máximo:", max_orig)
print("Datos normalizados:", datos_normalizados)
````

````python
# CUARTO CASO DE PRUEBA: desnormalizar la lista de datos normalizados
datos_originales = [40,60,80,100]
# Normalización al rango [-5, 5]
rang_inf = -5
rang_sup = 5
datos_normalizados, min_orig, max_orig = normalizar_lista(datos_originales, rang_inf, rang_sup)

print("Datos normalizados:", datos_normalizados)

# Recuperamos los datos originales
datos_recuperados = desnormalizar_lista(datos_normalizados, min_orig, max_orig, rang_inf, rang_sup)
print("Datos des-normalizados:", datos_recuperados)
````

Normalizar es llevar todos los datos a una misma escala, sin cambiar su relación entre ellos.

Normalizar los datos significa re-escalar los valores de una lista (o conjunto de datos) para que queden dentro de un rango fijo, por ejemplo [0,1] o [-1,1].

¿Por qué es útil normalizar?

Para que los datos no dependan de sus unidades (metros, grados, pesos, etc.).

Para que todos los valores tengan la misma escala.
